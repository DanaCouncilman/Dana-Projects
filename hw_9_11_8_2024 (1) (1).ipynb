```python

# Define the function
def find_unprecedented_sentences(sentences1, query_terms=None, similarity_threshold=0.2):
    """
    This function searches for phrases in the text that are similar to the words/phrases in 'query_terms'.
    """
    # Set default query terms if not provided
    if query_terms is None:
        query_terms = ["unprecedented", "never before", "unique", "extraordinary", "unlike anything"]

    # Initialize TF-IDF Vectorizer
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(sentences1)

    # List to store matching sentences
    matching_sentences = []

    # Loop through each text and check for similarity with query terms
    for i, text in enumerate(sentences1):
        for term in query_terms:
            term_vec = vectorizer.transform([term])
            similarity = cosine_similarity(term_vec, tfidf_matrix[i])
            if similarity[0][0] > similarity_threshold:
                matching_sentences.append((sentences1[i], similarity[0][0]))

    # Return the list of matching sentences
    return matching_sentences


# Run the function on preprocessed texts
results = find_unprecedented_sentences(sentences1)

# Display results
print("Sentences indicating acknowledgment of unprecedented conditions:")
for sentence, score in results:
    print(f"Score: {score:.2f} | Sentence: {sentence}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# Initialize VADER Sentiment Analyzer
sid = SentimentIntensityAnalyzer()

# List of texts 
texts = [text1, text2, text3, text4, text5]

# Meeting dates in order (for the x-axis)
meeting_dates = ["January", "March", "July", "September", "October"]

# Function to extract speaker and calculate sentiment for each meeting
def analyze_sentiment(texts, meeting_dates):
    '''analyzes the positive and negative sentiments in each of the five texts (not differentiated by speaker'''
    # List to hold sentiment data for all meetings
    sentiment_data = []

    for text_index, text in enumerate(texts):
        paragraphs = text.strip().split("\n")  # Split the text into paragraphs (assuming each paragraph is one speaker's turn)
        paragraphs = [p.strip() for p in paragraphs if p.strip()]  # Remove any empty lines or leading/trailing spaces

        total_sentiment_score = 0
        sentence_count = 0
        
        for paragraph in paragraphs:
            # Adjusted regex pattern for uppercase speaker names followed by a period (with optional spaces)
            match = re.match(r"([A-Z\s]+)\.\s*(.*)", paragraph.strip())  # Match uppercase speaker names followed by a period
            if match:
                content = match.group(2).strip()

                # Calculate sentiment score using VADER
                sentiment_score = sid.polarity_scores(content)["compound"]
                total_sentiment_score += sentiment_score
                sentence_count += 1
        
        # Calculate average sentiment for the meeting
        if sentence_count > 0:
            average_sentiment_score = total_sentiment_score / sentence_count
        else:
            average_sentiment_score = 0

        # Store the sentiment score along with the meeting date
        sentiment_data.append({
            'meeting_date': meeting_dates[text_index],
            'average_sentiment': average_sentiment_score
        })
    
    # Convert the sentiment data into a Pandas DataFrame
    sentiment = pd.DataFrame(sentiment_data)
    return sentiment

# Run sentiment analysis on the provided texts
sentiment = analyze_sentiment(texts, meeting_dates)
sentiment.head()
```

```python
# Plot the sentiment over time
plt.figure(figsize=(10, 6))

# Plot the average sentiment over the meetings
plt.plot(sentiment['meeting_date'], sentiment['average_sentiment'], marker='o', color='b', linestyle='-', linewidth=2, markersize=8)

# Add title and labels
plt.title("Average Sentiment Score of FOMC Meetings (2008)", fontsize=14)
plt.xlabel("Meeting Date", fontsize=12)
plt.ylabel("Average Sentiment Score (VADER)", fontsize=12)

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```

Step 5: Get Embeddings for Each Paragraph and Find Similarities
Compute embeddings for each paragraph and find the similarity between each paragraph’s embedding and the key phrase embeddings. Cosine similarity can be used for this.

```python
from sklearn.metrics.pairwise import cosine_similarity


def get_embedding(text, model="text-embedding-ada-002"):
    response = openai.embeddings.create(
        input=text,
        model=model
    )
    return response.data[0].embedding

# Get embeddings for each paragraph
paragraph_embeddings = [get_embedding(paragraph) for paragraph in paragraphs]

key_embeddings = {phrase: get_embedding(phrase) for phrase in key_phrases}

# Function to find similar paragraphs based on cosine similarity
def find_similar_paragraphs(paragraph_embeddings, key_embeddings, threshold=0.6):
    similar_paragraphs = []
    
    for i, paragraph_emb in enumerate(paragraph_embeddings):
        for phrase, key_emb in key_embeddings.items():
            similarity = cosine_similarity([paragraph_emb], [key_emb])[0][0]
            if similarity > threshold:
                similar_paragraphs.append((paragraphs[i], phrase, similarity))
                
    return similar_paragraphs

# Find paragraphs similar to the key phrases
results = find_similar_paragraphs(paragraph_embeddings, key_embeddings)
```

Step 6: Display Results
Print the paragraphs that are most similar to our key phrases, along with the similarity score.

```python
for paragraph, phrase, similarity in results:
    print(f"Similarity Score: {similarity:.2f}")
    print(f"Matched Key Phrase: {phrase}")
    print("Paragraph:", paragraph)
    print("-" * 80)
```
