```python
terrorism = pd.read_csv("rand_terrorism.csv", encoding="latin-1")
terrorism_keywords = {
    "economics_words": [
        "bank", "financial", "stock", "investment",  "credit",  "insurance", "ATM", "currency", "brokerage", "oil", "pipeline", "factory", "power plant", "trade center", "industrial park",
        "commercial building", "supply chain", "mining", "seaport","trade", "export", "import", "tariff", "market price", "sanctions",
        "commodity", "embargo", "economic blockade", "smuggling", "trade", "economic", "corporate", "business", "multinational", "enterprise", "economic stability", "inflation", "unemployment", "tax revenue", "gross domestic product",
        "economic reform", "financial crisis", "stock market crash", "recession", "interest rates", "tax", "taxes", "tax policy", "subsidies", "austerity", "economic sanctions", "regulatory policy", "trade agreement", "public finance", "economic stimulus", "budget cuts", "fiscal policy"],
    "religious_terminology": [
        "faith", "religion", "holy", "sacred", "martyr", "blasphemy", "heresy", "idol",
        "doctrine", "scripture", "pilgrimage", "worship", "sanctuary", "divine", "sacrilege", "prophet", "imam", "priest", "monk", "rabbi", "preacher", "clergy", "missionary",
        "pastor", "saint", "apostle", "disciple", "mullah", "ayatollah", "swami", "guru", "church", "mosque", "synagogue", "temple", "shrine", "monastery", "cathedral",
        "holy site", "sacred site", "basilica", "chapel",  "prayer", "fasting", "ritual", "sacrifice", "ceremony", "feast", "meditation", "penance", "pilgrimage", "confession", "holy day", "sabbath",
        "mass", "sermon", "bible", "quran", "torah", "vedas", "sutras", "scripture", "gospel", "talmud", "hadith", "bhagavad gita", "new testament", "old testament", "holy book"
        "christian", "muslim", "jewish", "hindu", "buddhist", "sikh", "catholic", "protestant", "orthodox", "sunni", "shia", "islamic", "zionist", "evangelical" "jihad", "crusade", "holy war", "fundamentalism", "radicalism", "extremism",
        "militant", "religious zealot", "martyrdom", "fatwa", "takfir", "apostate", "infidel", "holy"], 
      "government_words": ["government", "state", "regime", "administration", "dictator", "monarchy", "republic",
        "bureaucracy", "official", "public office", "executive order", "ministry", "parliament", "congress", "nationalism", "socialism", "communism",  "president", "prime minister", "chancellor", "king", "queen", "senator", "congressman",
        "parliamentarian", "mayor", "ambassador", "party leader", "party", "opposition leader", "protest", "demonstration", "uprising", "civil disobedience", "rebellion", "strike",
        "revolution", "insurgency", "resistance movement", "coup", "separatist movement", "activism", "freedom fighters", "policy", "legislation", "law", "reform", "bill", "constitutional", "amendment",
        "civil rights", "human rights", "law enforcement", "judiciary", "court", "supreme court", "international law", "war", "conflict", "military operation", "occupation", "invasion", "border dispute",
        "ceasefire", "treaty", "territorial dispute", "election", "ballot", "voting", "referendum", "poll", "campaign", "candidate", "electoral fraud", "vote rigging", "election observer", "election interference", "voter suppression", "liberation", "embassy" ] 
}


def categorize_incident(description):
    '''categorizes incidents based on above keywords'''
    if not isinstance(description, str):
        return "other"  # Return 'other' if the description is not a valid string

    # Create a dictionary to hold the category matches
    categories = {
        "economics": any(word in description.lower() for word in terrorism_keywords["economics_words"]),
        "religion": any(word in description.lower() for word in terrorism_keywords["religious_terminology"]),
        "politics": any(word in description.lower() for word in terrorism_keywords["government_words"])
    }

    # Collect true categories
    true_categories = [key for key, value in categories.items() if value]

    # Determine which category to return
    if len(true_categories) == 1:
        return true_categories[0]  # Only one category is true
    elif len(true_categories) > 1:
        return "combination"  # More than one category is true
    else:
        return "other"  # No categories are true


# Preprocess the description column
terrorism['Description'] = terrorism['Description'].fillna('').astype(str)

# Apply the categorization function
terrorism['category'] = terrorism['Description'].apply(categorize_incident)

# Save the updated DataFrame to a new CSV file
terrorism.to_csv('categorized_terrorism_incidents.csv', index=False)

# Display the counts of each category
category_counts = terrorism['category'].value_counts()
print("Incident counts by category:")
print(category_counts)

# Display the first few rows of the categorized data
print("\nSample categorized incidents:")
print(terrorism[['Description', 'category']].head())
 
```
