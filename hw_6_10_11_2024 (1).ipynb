```python
import re
from collections import Counter

# Use a correct file path (replace with your actual path)
repo_path = r'C:\Users\Dana Councilman\Downloads\speeches.txt'  
search_words = ['love', 'loved', 'I love']  # List of search words
context_words = 2  # Number of surrounding words to capture

def trump_loves(repo_path, search_words, context_words=2):
    '''This function tries to find what Trump loves (top 10) from the repository.'''
    
    # Combine search words into a single regex pattern (case-insensitive)
    search_pattern = '|'.join([re.escape(word) for word in search_words])  # Escape words for regex
    pattern = r'(?i)(\w+\s+){0,' + str(context_words) + r'}\b(?:' + search_pattern + r')\b(\s+\w+){0,' + str(context_words) + r'}'
    
    # Counter to hold surrounding words
    surrounding_word_counter = Counter()

    try:
        # Open the speech text file directly
        with open(repo_path, 'r', encoding='utf-8', errors='ignore') as file:
            content = file.read()

            matches = re.findall(pattern, content)
            print(f"Matches Found: {len(matches)}")  # Print the number of matches found
            print(f"Matches: {matches}")  # Print the actual matches

            for match in matches:
                # Get the surrounding words and update the counter
                surrounding_words = re.findall(r'\w+', ' '.join(match))
                surrounding_word_counter.update(surrounding_words)

    except Exception as e:
        print(f"Could not read {repo_path}: {e}")

    return surrounding_word_counter

# Find surrounding words
surrounding_words_count = trump_loves(repo_path, search_words, context_words)

# Print the most common surrounding words
if surrounding_words_count:
    for word, count in surrounding_words_count.most_common(10):  # Change 10 to however many you want to display
        print(f"'{word}' appears {count} times surrounding the words {', '.join(search_words)}.")
else:
    print(f"No surrounding words found for {', '.join(search_words)}.")

```

Write code that accepts the number as an input and then collects user input(in Bambara), translates it to French, and then outputs the data to a csv file in English. Please make sure you write code that addresses the challenges above (1-5).

```python
import csv
import random
import pandas as pd
from googletrans import Translator
import pandas as pd
import os

# Initialize the translator
translator = Translator()

# Keywords for routing calls
health_keywords = ["baby", "infant", "emergency", "hospital", "sick", "enfant"]
music_keywords = ["music", "song", "album", "studio", "Mango Mali"]
agriculture_keywords = ["mango", "price", "yield", "quality", "market", "production"]

# Simulate historical data for validation
historical_data = {
    'region1': {'yield': 500, 'price': 2.5},
    'region2': {'yield': 450, 'price': 2.0},
}

def save_to_csv(data, filename="agriculture_data.csv"):
    # Check if file exists before trying to append to it
    file_exists = os.path.isfile(filename)
    
    df = pd.DataFrame([data])
    
    # If the file doesn't exist, write with header; otherwise, append without header
    df.to_csv(filename, mode='a', header=not file_exists, index=False)
    print(f"Data saved to {filename}")


# Translate from Bambara to French, then French to English
def translate_bambara_to_english(text):
    try:
        # Bambara -> French
        french_translation = translator.translate(text, src='bm', dest='fr').text
        # French -> English
        english_translation = translator.translate(french_translation, src='fr', dest='en').text
        return english_translation
    except Exception as e:
        print(f"Translation error: {e}")
        return text  # fallback in case of error

# Main function to handle the call
def handle_call(phone_number, input_audio):
    print(f"Processing call from number: {phone_number}...")
    
    # Translate the audio from Bambara to English
    translated_text = translate_bambara_to_english(input_audio)
    print(f"Translated text: {translated_text}")
    
    # Identify the type of call
    call_type = identify_call_type(translated_text)
    
    # Route calls
    if call_type in ["health", "music"]:
        return redirect_call(call_type)
    
    # If it's an agriculture call, collect data
    if call_type == "agriculture":
        agricultural_data = collect_agriculture_data(translated_text)
        print(f"Collected agricultural data: {agricultural_data}")
        
        # Check for suspicious data
        if flag_suspicious_data(agricultural_data):
            return "Data appears suspicious. Please verify your input."
        else:
            # Save data to CSV
            save_to_csv(agricultural_data)
            return "Agricultural data recorded successfully."
    
    return "Call ended."

# Identify call type
def identify_call_type(translated_text):
    if any(word in translated_text for word in health_keywords):
        return "health"
    elif any(word in translated_text for word in music_keywords):
        return "music"
    elif any(word in translated_text for word in agriculture_keywords):
        return "agriculture"
    return "unknown"

# Collect agricultural data (price, yield, quality)
def collect_agriculture_data(translated_text):
    data = {}
    if "price" in translated_text:
        data["price"] = random.randint(1, 5)
    if "yield" in translated_text:
        data["yield"] = random.randint(100, 700)
    if "quality" in translated_text:
        data["quality"] = random.choice(["good", "average", "poor"])
    data["region"] = "region1"  # Default region for simplicity
    return data

# Redirect calls based on type
def redirect_call(call_type):
    if call_type == "health":
        return "Redirecting to infant health emergency hotline."
    elif call_type == "music":
        return "Redirecting to Mango Mali music studio."
    elif call_type == "agriculture":
        return "Proceeding with agricultural data collection."
    return "Unable to determine call type."

# Function to flag suspicious data
def flag_suspicious_data(data):
    region = data.get('region')
    yield_reported = data.get('yield')
    price_reported = data.get('price')
    historical_yield = historical_data[region]['yield']
    historical_price = historical_data[region]['price']
   
    if yield_reported is None:
        print(f"Error: yield_reported is None for region {region}")
        return False
    if price_reported is None:
        print(f"Error: price_reported is None for region {region}")
        return False
    
    if yield_reported < (0.7 * historical_yield) or price_reported < (0.7 * historical_price):
        return True
    return False



# Sample calls
sample_calls = [
    {"phone_number": "+22312345678", "input_audio": "prix mangue qualité production"},
    {"phone_number": "+22387654321", "input_audio": "urgence bébé hôpital"},
    {"phone_number": "+22311223344", "input_audio": "musique album Mango Mali"},
    {"phone_number": "+22344332211", "input_audio": "rendement mangue qualité"}
]

# Testing the system with sample calls
for call in sample_calls:
    result = handle_call(call["phone_number"], call["input_audio"])
    print(result)


```
